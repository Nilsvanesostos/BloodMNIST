```python
import medmnist
from medmnist import BloodMNIST
import numpy as np
import torch
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
```

```python
# Load dataset and apply transform (excluding normalization)
transform = transforms.Compose([
    transforms.ToTensor(),
])

# Load the dataset of the images
train_dataset = BloodMNIST(split='train', transform=transform, download=True, size=64) # Using size 64

# Create DataLoader for batch processing
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=False)
```

For medical images like BloodMNIST, the pixel values and distribution is different to the commonly used in CIFAR10. Let us compute the mean and standard deviation from the actual dataset:

```python
# Initialize mean and std
mean = 0.0
std = 0.0
total_images = 0

# Compute mean and std
for images, _ in train_loader:
    batch_samples = images.size(0)  # batch size
    images = images.view(batch_samples, images.size(1), -1)  # Flatten images
    mean += images.mean(2).sum(0)  # Calculate per channel mean
    std += images.std(2).sum(0)    # Calculate per channel std
    total_images += batch_samples

mean /= total_images
std /= total_images

print(f'Mean: {mean}')
print(f'Std: {std}')
```

```python
# Define the transform with the normalization using the values we just computed
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[mean[0], mean[1], mean[2]], std=[std[0], std[1], std[2]]),
])
```
